```python
# importing the libraries
import pandas as pd
import numpy as np
from deep_autoviml import deep_autoviml as deepauto
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
LLB = LabelEncoder()
```

```python
#Loading train and test dataset
datapath = 'https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv'
train = pd.read_csv(datapath, header=0, sep=',')

datapath = 'https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/test.csv'
test = pd.read_csv(datapath, header=0, sep=',')

train

```

```python
target = 'Survived'

keras_model_type = "auto"
project_name = "titanic-model"
model_options = {'nlp_char_limit':50, 'cat_feat_cross_flag':"",
                 'max_trials': 5, "tuner": "storm"}
keras_options = {"patience":10, 'class_weight': True, 'early_stopping': True, 
                 'lr_scheduler': '', "optimizer": 'RMS'}

model, vocab_dict = deepauto.fit(train, target, keras_model_type,project_name=project_name, keras_options=keras_options,
                                     model_options=model_options, save_model_flag=False, use_my_model='', verbose=1)
```

```python
##Inference on the test data
predictions = deepauto.predict(model, project_name, test_dataset=test, 
                               keras_model_type=keras_model_type, cat_vocab_dict=vocab_dict)
predictions

```
